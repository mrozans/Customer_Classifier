```python
import pandas as pd
import numpy as np
import DataLoad as data
```

```python
sessions_df = data.load_sessions_data()
users_df = data.load_users_data()
```

```python
cleared_sessions_df = sessions_df.loc[ (~sessions_df['user_id'].isna()) & (~sessions_df['product_id'].isna())]
cleared_sessions_df
```

```python
def get_offered_discounts(uid):
    return len(sessions_df.loc[(sessions_df['user_id'] == uid) & (sessions_df['offered_discount'] != 0)].index)

def get_buys_with_discount(uid):
    return len(sessions_df.loc[(sessions_df['user_id'] == uid) & (sessions_df['event_type_BUY_PRODUCT'] == 1) & (sessions_df['offered_discount'] != 0)].index)
```

```python
values = []
users = users_df['user_id']
for id in users:
    if  get_offered_discounts(id) < 2:
        values.append(0)
    else:
        values.append(get_buys_with_discount(id) / (get_offered_discounts(id) - 1))
values
```

```python
i = 0
selected_users = []
for id in users:
    if values[i] >= 0.1:
        selected_users.append(id)
    i = i + 1
selected_users
```
